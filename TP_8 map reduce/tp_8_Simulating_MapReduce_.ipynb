```python
from collections import defaultdict

# --- Mapper ---
def mapper(line):
    line = line.strip()
    if not line or line.startswith('#'):
        return []

    parts = line.split(',')
    if len(parts) < 7:
        return []

    status = parts[5].strip()  # extract status code
    if status == '200':  # Filter out successful responses
        return []
    return [(status, 1)]


def map_reduce(file_path):
    mapped = []

    # MAP PHASE
    with open(file_path, "r") as f:
        for line in f:
            mapped.extend(mapper(line))

    # SHUFFLE / GROUP PHASE
    grouped = defaultdict(list)
    for key, value in mapped:
        grouped[key].append(value)

    # REDUCE PHASE
    reduced = { key: sum(values) for key, values in grouped.items() }

    return reduced


results = map_reduce("weblogs.txt")

# --- OUTPUT ---
for status, count in sorted(results.items()):
    print(f"HTTP {status}: {count} requests")
```
